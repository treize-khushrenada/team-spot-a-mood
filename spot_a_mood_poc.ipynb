```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

# Two lists of sentences
dont_let_me_down_lyrics_sample = "I need you, I need you, I need you right now, Yeah, I need you right now, So don't let me, don't let me, don't let me down, I think I'm losing my mind now"

hello_lyrics_sample = "Hello, it's me, I was wondering if after all these years you'd like to meet, To go over everything, They say that time's supposed to heal ya, but I ain't done much healing"
```

```python
emb1 = model.encode(dont_let_me_down_lyrics_sample, convert_to_tensor=True)
emb2 = model.encode(hello_lyrics_sample, convert_to_tensor=True)
```

```python
cosine_scores = util.cos_sim(emb1, emb2)
```

```python
cosine_scores 
```

```python
user_input = "donald trump"
input_emb = model.encode(user_input, convert_to_tensor=True)
```

```python
cosine_score = util.cos_sim(input_emb , emb2)
cosine_score
```
